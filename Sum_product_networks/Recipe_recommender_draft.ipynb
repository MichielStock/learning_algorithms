```python
import theano
import theano.tensor as T
import numpy as np
import SPN
import pandas as pd
import random as rd
import matplotlib.pyplot as plt
from collections import deque
plt.ion()
%matplotlib inline
```

```python
def momentum(cost, params, learning_rate, momentum):
    """
    Stochastic gradient descent using momentum
    => ! minimiser !
    """
    grads = theano.grad(cost, params)
    updates = []
    for p, g in zip(params, grads):
        mparam_i = theano.shared(np.zeros(p.get_value().shape,
                    dtype=theano.config.floatX))
        v = momentum * mparam_i - learning_rate * g
        updates.append((mparam_i, v))
        updates.append((p, p + v))
    return updates
```

Here we will use a sum product network with a random topology to model recipes. The goal of this work is to use inference to complete, sample and study recipes.

```python
recipes = pd.DataFrame.from_csv('ReceptenBinair.csv', sep=';')
ingredients = recipes.columns
n_recipes, n_ingredients = recipes.shape
recipes
```

```python
# generate variables and marginal flags for each ingredient
variables = {ingr:(T.scalar(ingr), T.scalar(ingr+'_marg_flag')) for ingr in ingredients}
variables
```

```python
# five pdfs for each ingredient!
ingr_pdfs = {ingr : [SPN.BernoulliPDF(*variables[ingr]) for i in range(5)] for ingr in ingredients}
```

```python
# divide the ingredients in groups => apply domain knowledge HERE!
group_size = 5
ingr_list = list(ingredients)
rd.shuffle(ingr_list)
prod_layer1_ingr = [ingr_list[i*group_size:(i+1)*group_size] for i in range(n_ingredients/group_size + 1)]
prod_layer1_ingr
```

```python
# first product layer: 4 versions of each
product_layer1 = [
        [
            SPN.ProductNode([rd.choice(ingr_pdfs[ingr]) for ingr in combination], initial_layer=True) 
                   for i in range(5)] 
        for combination in prod_layer1_ingr]
```

```python
# sum layer 1
nsn1 = 12
np1 = len(product_layer1)
sum_layer1 = []

for prod_repetitions in product_layer1:
    combinations = []
    for rep in range(5):
        combinations.append(SPN.SumNode(prod_repetitions))
    sum_layer1.append(combinations)
```

```python
# product layer 2
number_product_nodes_l2 = 11
ns1 = len(sum_layer1)
product_layer2 = []
for combnr in range(number_product_nodes_l2):
    combinations = [SPN.ProductNode([rd.choice(sumnodes) for sumnodes in sum_layer1[combnr*ns1/number_product_nodes_l2:(combnr+1)*ns1/number_product_nodes_l2]]) for i in range(3)]
    product_layer2.append(combinations)
```

```python
# sum layer 2
nsn2 = 12
np2 = len(product_layer2)
sum_layer2 = []

for prod_repetitions in product_layer2:
    combinations = []
    for rep in range(5):
        combinations.append(SPN.SumNode(prod_repetitions))
    sum_layer2.append(combinations)
```

```python
# product layer 3
number_product_nodes_l3 = 4
ns2 = len(sum_layer2)
product_layer3 = []
for combnr in range(number_product_nodes_l3):
    combinations = [SPN.ProductNode([rd.choice(sumnodes)
                            for sumnodes in sum_layer2[combnr*ns2/number_product_nodes_l3:(combnr+1)*ns2/number_product_nodes_l3]]) 
                    for i in range(5)]
    product_layer3.append(combinations)
```

```python
# sum layer 3
np3 = len(product_layer3)
sum_layer3 = []

for prod_repetitions in product_layer3:
    combinations = []
    for rep in range(5):
        combinations.append(SPN.SumNode(prod_repetitions))
    sum_layer3.append(combinations)
```

```python
# final product layer
ns3 = len(sum_layer3)
product_layer4 = [SPN.ProductNode([rd.choice(sumnodes)
                        for sumnodes in sum_layer3]) 
                    for i in range(5)]
```

```python
# final sum node
final_sum_node = SPN.SumNode(product_layer4)
```

```python
parameters = final_sum_node.get_parameters()
learning_updates = momentum(-T.log(final_sum_node.max_output), parameters, 0.01, 0.8)  # negative, because we want to maximise
```

```python
inputs = [theano.Param(variables[ingr][0], name=ingr) for ingr in ingredients]  # ingredients
#inputs += [theano.Param(variables[ingr][1], default=1, name=ingr + '_flag') for ingr in ingredients]  # ingredients marginal flags => default is one
train_network = theano.function(inputs, final_sum_node.sum_output, updates=learning_updates)
```

##Model is defined, time to train!

```python
"""
plt.ion()
plt.xlabel('Training iteration')
plt.ylabel('Likelihood')
plt.grid()
plt.loglog()
"""
training_data = map(tuple, list(recipes.values))
rd.shuffle(training_data)


plot_nr = 0
likelihood = deque([])
for iteration in range(1000):
    for instance in training_data:
        likelihood.appendleft(train_network(*instance))
        if len(likelihood) > 500:
            likelihood.pop()
        if plot_nr % 500 == 0:
            print 'Iteration %s:'%plot_nr, np.mean(likelihood)
            #plt.scatter(plot_nr, np.mean(likelihood))
            #plt.show()
        plot_nr += 1
    rd.shuffle(training_data)
```
